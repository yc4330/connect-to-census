## 2-D visualizations (aka scatter plots)

It seems simple. It's not simple. mainly because...

**CORRELATION =/= CAUSATION**

Comparing variables to one another can be TRICKY!

Let's make some 2-D visualizations (aka scatter plots).

Be VERY careful how you phrase the headlines. Take time to think through what you want to say.




```python
%%R

# code for scatter plot 1
# be sure to include a meaningful title and subtitle
# scatter plot on count and population_estimate
# linear regression line
df %>%
  ggplot(aes(x = population_estimate, y= count)) +
  geom_point() +
  geom_smooth(method = "lm") +
  labs(title = "Number of illegal firework complaints vs Population",
       subtitle = "Each point represents a zipcode in New York City",
       y = "Number of illegal firework complaints")
```

```python
%%R

# code for scatter plot 2
# be sure to include a meaningful title and subtitle
# scatter plot on count and total_units_estimate
# linear regression line
df %>%
  ggplot(aes(x = total_units_estimate, y= count)) +
  geom_point() +
  geom_smooth(method = "lm") +
  labs(title = "Number of illegal firework complaints vs Housing unit",
       subtitle = "Each point represents a zipcode in New York City",
       y = "Number of illegal firework complaints")
```

```python
%%R

# code for scatter plot 3
# be sure to include a meaningful title and subtitle
# scatter plot on count and med_inc_estimate
# linear regression line
df %>%
  ggplot(aes(x = med_inc_estimate, y= count/population_estimate)) +
  geom_point() +
  geom_smooth(method = "lm") +
  labs(title = "Number of illegal firework complaints vs Median Income",
       subtitle = "Each point represents a zipcode in New York City",
       y = "Number of illegal firework complaints")
```
